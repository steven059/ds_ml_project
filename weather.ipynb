```python
#read file
_2019 = pd.read_csv("./일별/2019.csv",encoding='cp949')
_2020 = pd.read_csv("./일별/2020.csv",encoding='cp949')
_2021 = pd.read_csv("./일별/2021.csv",encoding='cp949')
_2022 = pd.read_csv("./일별/2022.csv",encoding='cp949')
_2023 = pd.read_csv("./일별/2023.csv",encoding='cp949')
```

```python
#drop col
_2019=_2019.drop(columns=drop_col)
_2020=_2020.drop(columns=drop_col)
_2021=_2021.drop(columns=drop_col)
_2022=_2022.drop(columns=drop_col)
_2023=_2023.drop(columns=drop_col)
```

```python
#concat
data = pd.concat([_2019,_2020,_2021,_2022,_2023],axis=0)
```

```python
data = data.reset_index(drop=True)
```

```python
data.info()
```

```python
data = data.fillna(0)
```

```python
data.to_csv('weather.csv',index='False')
```
