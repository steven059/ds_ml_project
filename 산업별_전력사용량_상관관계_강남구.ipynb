```python
import pandas as pd
import openpyxl
import numpy as np
```

```python
file_path = r'(2019~2023년)서울시 산업분류별 전력사용량.xlsx'

# Pu: Power usage
df_Pu = pd.read_excel(file_path, header=2)
```

```python
gangnam_df= df_Pu[df_Pu['시군구'] == '강남구']
```

```python
gangnam_df
```

```python
gangnam_df['일자'] = gangnam_df['일자'].apply(lambda x: x.replace(' 년 ', '').replace('월', ''))
```

```python
gangnam_df
```

```python
gangnam_df['사용량(kWh)'] = gangnam_df['사용량(kWh)'].replace('-', np.nan)
gangnam_df['사용량(kWh)'] = gangnam_df['사용량(kWh)'].str.replace(',', '').str.strip().replace(np.nan, '0').astype(int)
```

```python
import calendar
```

```python
def month_to_daily(row):
    year = int(row['일자'][:4])
    month = int(row['일자'][4:6])
    
    # 해당 월의 일수 계산
    num_days = calendar.monthrange(year, month)[1]
    
    # 월별 전력 사용량을 일수로 나눔
    daily_electricity = row['사용량(kWh)'] / num_days
    
    # 일별 데이터 생성
    daily_data = []
    for day in range(1, num_days + 1):
        daily_data.append({
            '일자': f"{year}{month:02d}{day:02d}",  # YYYYMMDD 형식으로 날짜 생성
            '시군구': row['시군구'],
            '계약구분': row['계약구분'],
            '사용량(kWh)': daily_electricity
        })
    
    return daily_data

# 모든 행에 대해 일별 데이터 생성
all_daily_data = []
for index, row in gangnam_df.iterrows():
    daily_data = month_to_daily(row)
    all_daily_data.extend(daily_data)

# 일별 데이터로 DataFrame 생성
daily_df = pd.DataFrame(all_daily_data)

```

```python
merged_df = pd.merge(daily_df, gangnam_air, on=['지역','일자'], how='left')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

grouped = merged_df.groupby('계약구분')

# 상관관계 계산 및 결과 저장
correlation_results = {}

for name, group in grouped:
    corr = group[['사용량(kWh)', 'SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25']].corr()
    correlation_results[name] = corr

# 결과 시각화
for name, corr in correlation_results.items():
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Correlation Matrix for 계약구분: {name}')
    plt.show()
```
